# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Column pairs to visualize
columns_to_plot = [('Detergents_Paper','Milk'), 
                   ('Grocery','Fresh'), 
                   ('Frozen','Delicassen')]

# Number of clusters to experiment with
cluster_options = [2, 3, 4, 5]

# Loop over each column pair and cluster number
for x_col, y_col in columns_to_plot:
    plt.figure(figsize=(20,4))
    for i, n_clusters in enumerate(cluster_options):
        # Fit K-Means
        kmeans = KMeans(n_clusters=n_clusters, random_state=42)
        labels = kmeans.fit_predict(customers_scale)
        
        # Temporary DataFrame for plotting
        df_temp = df.copy()
        df_temp['labels_temp'] = labels
        
        # Plot side by side
        plt.subplot(1, len(cluster_options), i+1)
        sns.scatterplot(data=df_temp, x=x_col, y=y_col, hue='labels_temp', palette='Set1', legend=False)
        plt.title(f'{n_clusters} clusters')
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        
    plt.suptitle(f'{x_col} vs {y_col} - K-Means Cluster Comparison', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

**Your comment here**
Conclusion:

The most appropriate number of clusters appears to be K = 4 or K = 5, as these values provide the best balance between interpretability and separation of customer groups.


# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# Column pair to visualize
x_col, y_col = 'Detergents_Paper', 'Milk'

# List of parameter combinations to try
dbscan_params = [
    {'eps':0.3, 'min_samples':3},
    {'eps':0.5, 'min_samples':5},
    {'eps':0.7, 'min_samples':5},
    {'eps':1.0, 'min_samples':10}
]

# Loop through each combination and visualize
plt.figure(figsize=(20,4))
for i, params in enumerate(dbscan_params):
    # Fit DBSCAN
    dbscan = DBSCAN(eps=params['eps'], min_samples=params['min_samples'])
    labels = dbscan.fit_predict(customers_scale)
    
    # Temporary DataFrame for plotting
    df_temp = df.copy()
    df_temp['labels_temp'] = labels
    
    # Scatter plot
    plt.subplot(1, len(dbscan_params), i+1)
    sns.scatterplot(data=df_temp, x=x_col, y=y_col, hue='labels_temp', palette='Set2', legend=False)
    plt.title(f"eps={params['eps']}, min_samples={params['min_samples']}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    
plt.suptitle(f'DBSCAN Clustering Variations ({x_col} vs {y_col})', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**Your comment here**
The performance of DBSCAN is highly sensitive to parameter selection. In this dataset:

Moderate values of eps (around 0.5) provide the most interpretable clustering.
min_samples = 5 appears to give a reasonable balance between noise detection and cluster formation.

However, compared to K-Means, DBSCAN produces less stable and less interpretable clusters for this dataset.
